<h4 style="color:brown">Problem: predict the salary based on the number of years of experience.</h4>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("Salary.csv")
df
```

```python
%matplotlib inline
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df.YearsExperience,df.Salary,color='red',marker='+')
```

```python
new_df = df.drop('Salary',axis='columns')
new_df.head()
```

```python
Salary=df.Salary
Salary
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,Salary)
```

##### Predict Salary of the employee with years of experience =15

```python
reg.predict([[15]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Y=mX+C
15*8731.94104706+28859.52619453144
```

##### Predict Salary of the employee with years of experience =10

```python
reg.predict([[10]])
```

### Generate CSV file with list of Salary Prediction

```python
YOE_df = pd.read_csv("YOE.csv")
YOE_df.head()
```

```python
p = reg.predict(YOE_df)
p
```

```python
YOE_df['Salary']=p
YOE_df
```

```python
YOE_df.to_csv("Predicted Salary.csv")
```
